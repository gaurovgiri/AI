### Graph Coloring with NetworkX

```python
colors = ['red', 'blue','green', 'yellow', 'pink']
states = ['WA','NT','SA', 'Q','NSW', 'V', 'T']
```

```python
neighbours = dict()
neighbours["WA"] = ["NT","SA"]
neighbours["NT"] = ["WA", "SA", "Q"]
neighbours["SA"] = ["WA", "NT", "Q", "NSW", "V"]
neighbours["Q"] = ["NT", "SA", "NSW"]
neighbours["NSW"] = ["Q","SA","V"]
neighbours["V"] = ["SA", "NSW"]
neighbours["T"] = []
print(neighbours)
```

```python
color_of_states = dict()

for state in states:
    for color in colors:
        for neighbour in neighbours[state]:
            if color_of_states.get(neighbour) == color:
                break
        else:
            color_of_states[state] = color
            break

print(color_of_states)
```

```python
import networkx


graph = networkx.Graph()
graph.add_nodes_from(states)
graph.nodes()

graph.add_edges_from([('WA', 'NT'), ('WA', 'SA'), ('NT', 'SA'), ('NT', 'Q'), ('SA', 'Q'), ('SA', 'NSW'), ('SA', 'V'), ('Q', 'NSW'), ('NSW', 'V')])
graph.edges()

networkx.draw(graph, with_labels=True, node_color=color_of_states.values(),node_size=1000)
```
